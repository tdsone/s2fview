# Interactive `s2fview` viewer

IGV-style interactions wired into our matplotlib viewer via the `ipympl`
backend. **Run this notebook top-to-bottom in a live Jupyter / JupyterLab /
VS Code session** — the widget magic *must* be the first thing executed, so
don't paste these cells into a notebook that already drew inline figures.

If the canvas appears static and unresponsive:

1. *Restart the kernel.*
2. Re-run all cells from the top — `%matplotlib widget` must run before any
   `coverage_track(...)` call in this kernel session.

Once active:

- **Click + drag** on the coverage to zoom into that x-range.
- **Double-click** anywhere on a track to reset the zoom.
- **Hover** to see a thin crosshair + position readout across both tracks.

```python
%matplotlib widget
```

```python
import numpy as np

from s2fview import Gene, interactive_coverage_track
```

```python
rng = np.random.default_rng(0)
length = 500
positions = np.arange(length)

signal = 20 + 15 * np.exp(-((positions - 150) ** 2) / (2 * 30**2))
signal += 25 * np.exp(-((positions - 350) ** 2) / (2 * 50**2))
coverage = np.maximum(0, signal + rng.normal(0, 2, size=length)).round().astype(int)

sequence = "".join(rng.choice(list("ACGT"), length))

genes = [
    Gene(
        name="BIG1",
        start=20,
        end=380,
        strand="+",
        exons=((20, 55), (110, 160), (220, 270), (340, 380)),
    ),
    Gene(
        name="anti-BIG",
        start=80,
        end=300,
        strand="-",
        exons=((80, 130), (200, 240), (280, 300)),
    ),
    Gene(name="tiny", start=190, end=205, strand="+"),
    Gene(
        name="TAIL",
        start=350,
        end=485,
        strand="-",
        exons=((350, 390), (430, 485)),
    ),
    Gene(name="upstream-RNA", start=5, end=45, strand="-"),
]
```

```python
interactive_coverage_track(
    coverage,
    positions,
    hover_hz=10,
    genes=genes,
    sequence=sequence,
    title="Drag to zoom (letters appear once there's room) · dbl-click to reset",
)
```

```python
interactive_coverage_track(
    coverage,
    positions,
    hover_hz=60,
    genes=genes,
    sequence=sequence,
    title="Drag to zoom (letters appear once there's room) · dbl-click to reset",
)
```

```python
interactive_coverage_track(
    coverage,
    positions,
    hover_hz=1,
    genes=genes,
    sequence=sequence,
    title="Drag to zoom (letters appear once there's room) · dbl-click to reset",
)
```
