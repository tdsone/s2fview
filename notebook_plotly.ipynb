# Plotly viewer (parallel to `notebook_interactive.ipynb`)

`s2fview.plotly_track.plotly_coverage_track` mirrors the matplotlib viewer
but renders **entirely in the browser** via Plotly. Hover (spike-line
crosshair), pan, zoom, reset, and box-select all run client-side as native
JS — there is no Python round-trip per interaction, so there's nothing
that *can* lag.

Trade-offs vs. the matplotlib version:

| Aspect | matplotlib + ipympl | Plotly |
|---|---|---|
| Hover crosshair | Python round-trip per move (throttled) | Native JS, zero-latency |
| Static export | Pixel-perfect, integrates with mpl pipeline | Static PNG / SVG via kaleido (extra dep) |
| Styling control | Full Python-level over every artist | Fluent but more constrained |
| First display | Fast | Pulls plotly.js bundle |

Open the notebook in a live JupyterLab / VS Code session — Plotly figures
also render in static exports (the JSON state is embedded in the cell
output, so nbconvert HTML works too).

```python
import numpy as np

from s2fview import Gene
from s2fview.plotly_track import plotly_coverage_track

rng = np.random.default_rng(0)
length = 500
positions = np.arange(length)

signal = 20 + 15 * np.exp(-((positions - 150) ** 2) / (2 * 30**2))
signal += 25 * np.exp(-((positions - 350) ** 2) / (2 * 50**2))
coverage = np.maximum(0, signal + rng.normal(0, 2, size=length)).round().astype(int)

sequence = "".join(rng.choice(list("ACGT"), length))

genes = [
    Gene(
        name="BIG1",
        start=20,
        end=380,
        strand="+",
        exons=((20, 55), (110, 160), (220, 270), (340, 380)),
    ),
    Gene(
        name="anti-BIG",
        start=80,
        end=300,
        strand="-",
        exons=((80, 130), (200, 240), (280, 300)),
    ),
    Gene(name="tiny", start=190, end=205, strand="+"),
    Gene(
        name="TAIL",
        start=350,
        end=485,
        strand="-",
        exons=((350, 390), (430, 485)),
    ),
    Gene(name="upstream-RNA", start=5, end=45, strand="-"),
]
```

```python
plotly_coverage_track(
    coverage,
    positions,
    genes=genes,
    sequence=sequence,
    title="Plotly viewer — hover/zoom/pan run client-side, no lag",
)
```
